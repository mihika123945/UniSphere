```python
import requests
import pandas as pd
import time

# STEP 1: Set your API Key
API_KEY = "qpbwlxl4iTlOgHi0oBv2LHnpJqrlvI50yQeTTbAU"  # Replace with your real API key from https://api.data.gov/signup

# STEP 2: Define base URL and fields
BASE_URL = "https://api.data.gov/ed/collegescorecard/v1/schools"
FIELDS = [
    "school.name",
    "school.city",
    "school.state",
    "school.country",
    "school.school_url",
    "latest.cost.tuition.in_state",
    "latest.cost.tuition.out_of_state",
    "latest.admissions.admission_rate.overall",
    "latest.student.size",
    "latest.programs.cip_4_digit.title"
]

# STEP 3: Prepare the request template
params = {
    "api_key": API_KEY,
    "fields": ",".join(FIELDS),
    "per_page": 100  # max allowed per page
}

# STEP 4: Make paginated requests
all_results = []
total_pages = 60  # ~6000 records (100 per page * 60 pages)

for page in range(total_pages):
    print(f"Fetching page {page+1} of {total_pages}...")
    params["page"] = page
    response = requests.get(BASE_URL, params=params)
    
    if response.status_code == 200:
        all_results.extend(response.json()["results"])
    else:
        print(f"Error fetching page {page}: {response.status_code}")
        break  # stop if we hit an error

    time.sleep(0.5)  # be polite to the API

# STEP 5: Convert to DataFrame and Save
df = pd.DataFrame(all_results)
df.to_csv("us_universities_full_data.csv", index=False)
print("✅ Data saved to us_universities_full_data.csv")
```

```python
import pandas as pd

# Load university dataset
univ_df = pd.read_csv("us_universities_full_data.csv")

# Load cost of living dataset
cost_df = pd.read_csv("advisorsmith_cost_of_living_index.csv")
```

```python
# Clean column names (only if they look messy or inconsistent)
univ_df.columns = univ_df.columns.str.strip().str.lower()
cost_df.columns = cost_df.columns.str.strip().str.lower()
```

```python
print(cost_df.columns.tolist())
```

```python
# Step 1: Clean column names
cost_df.columns = cost_df.columns.str.strip().str.lower().str.replace(" ", "_")

# Step 2: Define baseline and calculate monthly cost
baseline = 1500  # Index 100 = $1500/month
cost_df["avg_monthly_cost_usd"] = round((cost_df["cost_of_living_index"] / 100) * baseline, 2)

# Step 3: Confirm column added
print(cost_df.head())
```

```python
# For universities
univ_df["school.city"] = univ_df["school.city"].str.strip().str.lower()
univ_df["school.state"] = univ_df["school.state"].str.strip().str.upper()

# For cost of living
cost_df["city"] = cost_df["city"].str.strip().str.lower()
cost_df["state"] = cost_df["state"].str.strip().str.upper()
```

```python
merged_df = univ_df.merge(
    cost_df[["city", "state", "avg_monthly_cost_usd"]],
    left_on=["school.city", "school.state"],
    right_on=["city", "state"],
    how="left"
)
```

```python
merged_df.to_csv("merged_data.csv", index=False)
```

```python
import pandas as pd

# Step 1: Load both datasets
univ_df = pd.read_csv("merged_data.csv")
geo_df = pd.read_csv("uscities.csv")

# Step 2: Keep only relevant columns from geolocation file
geo_df = geo_df[["city", "state_id", "lat", "lng"]]
geo_df.columns = ["city", "state", "latitude", "longitude"]

# Step 3: Clean and standardize both datasets for merging
geo_df["city"] = geo_df["city"].str.strip().str.lower()
geo_df["state"] = geo_df["state"].str.strip().str.upper()

univ_df["school.city"] = univ_df["school.city"].str.strip().str.lower()
univ_df["school.state"] = univ_df["school.state"].str.strip().str.upper()

# Step 4: Merge on city and state
merged_with_coords = univ_df.merge(
    geo_df,
    left_on=["school.city", "school.state"],
    right_on=["city", "state"],
    how="left"
)

# Step 5: (Optional) Save the result
merged_with_coords.to_csv("merged_data_with_coordinates.csv", index=False)

# Step 6: Preview
print(merged_with_coords[["school.name", "school.city", "school.state", "latitude", "longitude"]].head())
```
